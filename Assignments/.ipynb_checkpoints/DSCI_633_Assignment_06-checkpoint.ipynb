## Step 1: Download iris dataset and usual libraries


```python
import sys
import os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import mean_squared_error, accuracy_score
from math import sqrt
import seaborn as sns
from sklearn.datasets import load_iris
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from scipy.stats import mode
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading  the data
data = load_iris()
print(data)
```

We can see that the data is in the form of a dictionary.

```python
# Let's convert the data dictionary into a dataframe

# Get the features from the dictionary
features = data['data']

# Get the target from the dictionary
target = data['target']

# Create a dataframe of features
df_features = pd.DataFrame(features, columns=data['feature_names'])

# Create a dataframe of the target column
df_target = pd.DataFrame(target)

# Append the target column to the features dataframe to form the main dataframe
df_features['target'] = df_target


df = df_features
df.head()
```

## Step 2: Use GMM to separate the clusters using all 4 features: petal length & width, and sepal length & width. Map each cluster to a class by picking the most common class for each cluster (using the scipy.stats.mode() function): 

```python
# Let's convert all the features from the iris dataset into a numpy array
X = df.loc[:, df.columns!="target"].values

# Also let's store the target in y in the form of a numpy array
y = np.array(df.loc[:, df.columns=="target"])
```

```python
# Let us use a Gaussian Mixture Model to seperate the data into 3 clusters, since we know that the iris dataset has 3 categories - setosa, virginica, versicolor
# Using a random state of 18 since the cluster labels outputted using this random state are in the same order of the actual labels i.e. - 0, 1, 2 
gaussian_mixture = GaussianMixture(n_components=3, random_state=18)

# Fitting on only the features using a GMM model
y_pred = gaussian_mixture.fit_predict(X)


# Reshaping y into a one-dimensional array so that it has the same shape as y_pred
y = y.reshape(-1)
```

```python
# Actual labels
y
```

```python
ax = plt.figure()
plt.title("Plotting Sepal length vs Sepal Width using actual target values")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=y);
```

```python
# Predicted clusters
y_pred
```

As we can see, there are three clusters that have been formed and the cluster labels are also in the same order as those of the class labels i.e - 0, 1, 2. This is because of choosing the `random_state=18`. 

```python
ax = plt.figure()
plt.title("Plotting Sepal length vs Sepal Width using predicted Clusters")
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=y_pred);
```

From the above two scatterplots, we can observe that they are almost similar, except for a few obeservations in the cluster 1 that have been classified as belonging to cluster 2

In order to map each cluster to a class by picking the most common class for each cluster (using the scipy.stats.mode() function), we need to find the indices of the instances belonging to targets 0, 1, 2 in the original data.

```python
# class 0 index
class_0_indices = df[df['target']==0].index

# class 1 index
class_1_indices = df[df['target']==1].index

# class 2 index
class_2_indices = df[df['target']==2].index
```

We can see that all the observations for class 0 have been put into one cluster having the cluster label 0. We can map all the predicted clusters in the cluster label 0 to the class 0 in the target i.e. `setosa`.

```python
# Checking the predicted cluster labels for all the instances belonging to class 1
y_pred[class_1_indices]
```

We can see that some observations belonging to class 1 have been put into another cluster with the cluster label 2. Here we have to use the `scipy.stats.mode()` to map these observations to the correct class

```python
mode(y_pred[class_1_indices]).mode[0]
```

We can map all the predicted clusters in the cluster label 1 to the class 1 in the target i.e. `versicolor`.

```python
# Checking the predicted cluster labels for all the instances belonging to class 2
print(y_pred[class_2_indices])
print("Mode of cluster 2 is: ",mode(y_pred[class_2_indices]).mode[0])
```

We can see that all the observations for class 2 have been put into one cluster having the cluster label 2. 
We can map all the predicted clusters in the cluster label 2 to the class 2 in the target i.e. `virginica`.

## Step 4: Generate blobs from a new dataset, use “ from sklearn.datasets import make_blobs” . Use the following parameters:
- n_samples=2000

- blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],

- blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1])

```python
# Using make_blobs() to generate artificial features and target
features, target = make_blobs(n_samples=2000, centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3]]),
    cluster_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1]), random_state=18)



# Features
print("Features")
print(features)

print()

# Labels
print("Target_labels")
print(target)
```

```python
set(target)
```

As we can see `make_blobs()` has created 5 categories of target since we have provided 5 centers and 5 values for standard deviations.

## Step 5: Plot the blobs

```python
# Lets plot a scatter plot of the features, with the target as the hue
plt.scatter(features[:, 0], features[:, 1], marker='8', c=target,
            s=25, edgecolor="k");
```

## Step 6: Train k-means to train on this dataset

```python
# Let's instantiate a KMeans() model instance with 5 clusters initially, since we know that there are 5 categories in target 
model = KMeans(n_clusters=5, random_state=10)
model.fit(features)

clusters = model.predict(features)
```

## Step 7: Print the value of the centroids 

```python
# Lets now plot a scatter plot of the features with the predicted cluster labels as the hue
plt.scatter(features[:, 0], features[:, 1], marker='8', c=clusters,
            s=25, edgecolor="k");

# Getting the values of the cluster centroids
cluster_centroids = model.cluster_centers_
print("the cluster centorids are: ")

# Print the cluster label along with their centroids using for loop since the labels for the cluster centres are ordered
# Using the link https://stackoverflow.com/questions/29460891/labels-for-cluster-centers-in-python-sklearn as a reference.
for i in range(5):
    print("Cluster ", i, " has coordinate: ", cluster_centroids[i])

# Plotting the cluster centroids in the same scatterplot for clusters
plt.scatter(cluster_centroids[:,0] , cluster_centroids[:,1] , s = 80, color = 'r');
```

From the above plot and the cluster centroids, we can see that:
    - The purple cluster is cluster 0
    - The largest cluster i.e the blue cluster is cluster 1
    - The dark green cluster is cluster 2
    - The second largest cluster i.e. the light green cluster is cluster 3
    - The yellow cluster is the cluster 4

```python
X_new = np.array([[0, 2], [3, 2], [-3, 3], [-3, 2.5]])
y_pred = model.predict(X_new)
print("The predicted clusters for X_new are: ", y_pred)
```

## Step 9: Plot model’s decision boundaries using data, centroids,decision boundaries. Result should be a voronoi_plot  = 2 points 

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

plt.rcParams["figure.figsize"] = [15, 10]
plt.rcParams["figure.autolayout"] = True
ax = plt.subplot(211, aspect="equal")

#Plots Voronoi diagram of k-means clusters overlaid with data
def kmeans_voronoi_plot(X, y, cluster_centers):
    plt.scatter(X[:, 0], X[:, 1], marker='8', c=clusters, s=10, edgecolor="k")
    vor = Voronoi(cluster_centers)
    voronoi_plot_2d(vor, ax=ax, show_vertices=True, alpha=0.5, color="r", point_size=15)
    
kmeans_voronoi_plot(features, clusters, cluster_centroids)
```

## Step 10: Set values of k = 3 and k = 8, explain your results in both cases by means of the resultant intertia = 2 points 

```python
# Set the number of cluster equal to 3
model_3 = KMeans(n_clusters=3)
y_pred_3 = model_3.fit_predict(features)

# Set the number of cluster equal to 8
model_8 = KMeans(n_clusters=8)
y_pred_8 = model_8.fit_predict(features)

print(model_3.inertia_)
print(model_8.inertia_)
```

The inertia for every value of `k` no of clusters corresponds to the Within Cluster Sum of Squares (WCSS) value for that many clusters. As we increase the value for `k` it can be seen that the inertia reduces. That means, within every cluster, the distance between the data points of that cluster and their respective cluster centres reduces. 
Hence, we can say that as we increase the value of `k` the intra-cluster distances (i.e. distances of all points in a cluster from their centroids) decreases and the inter-cluster distances (i.e the distance between each of the k clusters/cluster centroids) increases

## Step 11: plot the inertia as a function of k and analyze the resulting curve, show plots, = 2 points

```python
inertia = []

# Lets check for cluster values of k from 3 to 10
k = [i for i in range(3, 11)]


for i in k:
    model = KMeans(n_clusters=i, random_state=18)
    model.fit_predict(features)
    inertia.append(model.inertia_)
    
plt.title("Inertia as a function of k")    
plt.xlabel("No of clusters (k)")
plt.ylabel("Inertia")
plt.plot(k, inertia);
```

From the above curve, we can see that there is a steep drop in the inertia at the k value of 4 beyond which the inertia decreases gradually. Hence this value should be chosen as the elbow point or the optimum value for the number of clusters since we can say that at this value, the data instances have been appropraitely grouped into the optimum number of clusters

## Step 11: what is the value of k at elbow

At the elbow point, the x-coordinate is approximately 4. So, the optimum number of clusters k that there can be is equal to 4.
